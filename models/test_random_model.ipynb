```python
import sys
import os
sys.path.append(os.path.join(os.path.abspath(''), '../'))

from models.random_model import RandomModel
from dataset.data_loader import DataLoader
from utils.utils import get_game_name_and_scores
```

```python
data_loader = DataLoader(cache_local_dataset=True, user_game_edge_embeddings=['playtime_forever'], game_embeddings=['name'])
data_loader.load_random_train_test_split(train_percentage=0.9, test_percentage=0.1, seed=0)
data_loader.save_data_loader_parameters('test_random_model_data_loader', overwrite=True)
```
