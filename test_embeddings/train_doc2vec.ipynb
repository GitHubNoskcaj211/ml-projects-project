```python
def process_customdoc2vec_embedding(desc, model):
    if desc is None:
        return None
    try:
        embedding = process_description(desc, lambda x: predict_from_model(model, x))
        return [item for sublist in embedding if sublist is not None for item in (sublist if isinstance(sublist, list) else [sublist])]
    except Exception as e:
        print(f"Error processing description: {desc} with model {model}. Error: {e}")
        return None

loaded_model = Doc2Vec.load('game_description.d2v')
custom_doc2vec_embeddings = [process_customdoc2vec_embedding(desc, loaded_model) for desc in df['description']]
custom_doc2vec_embeddings = [list(flatten(emb)) if emb is not None else None for emb in custom_doc2vec_embeddings]
# df['custom_doc2vec_embeddings'] = custom_doc2vec_embeddings
```

```python
def process_bert_embedding(desc):
    if desc is None:
        return None
    try:
        embedding = generate_bert_embedding(desc)
        return [item for sublist in embedding for item in sublist]
    except Exception as e:
        print(f"Error processing description: {desc}. Error: {e}")
        return None

bert_results = [process_bert_embedding(desc) for desc in df['description']]
# df['Bert_embeddings'] = bert_results
```
