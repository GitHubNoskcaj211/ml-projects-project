```python
import sys
import os
sys.path.append(os.path.join(os.path.abspath(''), '../'))

from evaluation.evaluation_harness import SAVED_EVALUATION_PATH
import os
import pandas as pd
```

```python
def parse_metrics_file(file_path):
    metrics = {}
    with open(file_path, 'r') as file:
        for line in file:
            key, value = line.strip().split(': ')
            metrics[key] = float(value)
    return metrics

subfolders = [f.path for f in os.scandir(SAVED_EVALUATION_PATH) if f.is_dir()]
data = []
for subfolder in subfolders:
    metrics_file_path = os.path.join(subfolder, 'metrics.txt')
    
    if os.path.exists(metrics_file_path):
        metrics_data = parse_metrics_file(metrics_file_path)
        data.append({'evaluation_name': subfolder.split('/')[-1], **metrics_data})

results = pd.DataFrame(data)
```

```python
def display_ranked_column(column_name):
    display(results.sort_values(by=column_name, ascending=False)[['evaluation_name', column_name]])
```
